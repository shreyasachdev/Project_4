```python
import boto3
import os
```

```python
s3 = boto3.resource('s3', 'us-east-1')
```

```python
image_file_paths = os.listdir('car_images')
```

```python
full_image_file_paths = []
for file in image_file_paths:
    full_image_file_paths.append(f"car_images/{file}")
full_image_file_paths  
```

```python
for file in full_image_file_paths:
    s3.meta.client.upload_file(file,'car-prediction-training-images',file)
```
